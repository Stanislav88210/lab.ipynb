```python
# считаем среднее количество лет работы в компании для сотрудников с рейтингом 2
employee_f[employee_f['rating'] == 2].mean()
```

```python
# считаем максимальный рейтинг сотрудников со стажем в 2 года
employee_f[(employee_f['years_of_working'] == 2)]['rating'].max()
```

```python
# среднее количество лет работы сотрудников для каждого рейтинга
employee_f.pivot_table(['years_of_working'], ['rating'], aggfunc='mean').head(10)
```

```python
# средний чек по продажам
orders_f['check_sum'].mean()
```

```python
# максимальный чек
orders_f['check_sum'].max()
```

```python
# среднее количество каждого сотрудника (средний чек-mean)
columns_to_show = ['Id','employee_id', 'check_sum']
orders_f.groupby(['employee_id'])[columns_to_show].describe(percentiles=[])
```

```python
# среднее количество продаж каждого клиента (средний чек-mean)
columns_to_show = ['Id','client_id', 'check_sum']
orders_f.groupby(['client_id'])[columns_to_show].describe(percentiles=[])
```

```python
# среднее количество продаж каждого  товара (средний чек-mean) с временем выставления инвойса
columns_to_show = ['Id','goods_id', 'check_sum', 'days_invoice']
orders_f.groupby(['client_id'])[columns_to_show].describe(percentiles=[])
```
