```python
data['diagnosis'][data['diagnosis'] == 'M']=0
data['diagnosis'][data['diagnosis'] == 'B']=1
```

```python
data.head()
```

```python
data[['diagnosis']] = data[['diagnosis']].astype('int64') 
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(data)
X_scaled = scaler.transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
print("Форма исходного массива: {}".format(str(X_scaled.shape)))
print("Форма массива после сокращения размерности: {}".format(str(X_pca.shape)))
```

```python
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], data.diagnosis)
plt.legend(data.columns, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("Первая главная компонента")
plt.ylabel("Вторая главная компонента")
```

```python
print("форма главных компонент: {}".format(pca.components_.shape))
```

```python
print("компоненты PCA:\n{}".format(pca.components_))
```

```python
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["Первая компонента", "Вторая компонента"])
plt.colorbar()
plt.xticks(range(len(data.columns)),
data.columns, rotation=60, ha='left')
plt.xlabel("Характеристика")
plt.ylabel("Главные компоненты")
```

```python
X_pca_tr=pd.DataFrame(X_pca)
X = X_pca_tr.values
y = data['diagnosis'].values
print("форма массива X: {} форма массива y: {}".format(X.shape, y.shape))
```

```python
#модель по исходным данным-низкое качество
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(data, data[['diagnosis']], test_size = 0.3, random_state = 0)
model = linear_model.SGDClassifier(alpha=0.0001, n_iter=100, random_state = 0)
model.fit(train_data, train_labels)
model_predictions = model.predict(test_data)
print (metrics.accuracy_score(test_labels, model_predictions))
print (metrics.classification_report(test_labels, model_predictions))
```

```python
# модель с использованием PCA
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(X_pca_tr, data[['diagnosis']], test_size = 0.3, random_state = 0)
model = linear_model.SGDClassifier(alpha=0.0001, n_iter=1000, random_state = 0)
model.fit(train_data, train_labels)
model_predictions = model.predict(test_data)
print (metrics.accuracy_score(test_labels, model_predictions))
print (metrics.classification_report(test_labels, model_predictions))
```

```python
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(X_pca_tr, data['diagnosis'], test_size = 0.1)
model = linear_model.SGDClassifier(alpha=0.001, n_iter=100, random_state = 0)
scores = cross_validation.cross_val_score(model, train_data, train_labels, cv=10)
print (scores.mean())
```

```python
# логистическая регрессия
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import LinearSVC

logreg = LogisticRegression().fit(train_data, train_labels)
print("Правильность на обучающем наборе: {:.3f}".format(logreg.score(train_data, train_labels))) 
print("Правильность на тестовом наборе: {:.3f}".format(logreg.score(test_data, test_labels)))
```

```python
# решающие деревья
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=0)
tree.fit(train_data, train_labels)
print("Правильность на обучающем наборе: {:.3f}".format(tree.score(train_data, train_labels))) 
print("Правильность на тестовом наборе: {:.3f}".format(tree.score(test_data, test_labels)))
```
