```python
proline_1 = [float(w) for w in wines_1['Proline']]
```

```python
def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums)
    for n in nums:
        diffs += (n - avg)**(2)
    return (diffs/(len(nums)-1))**(0.5)
```

```python
print(stdev(proline_1))
```

```python
plt.plot(proline_1)
```

```python
kurtosis(proline_1)
```

```python
skew(proline_1)
```
