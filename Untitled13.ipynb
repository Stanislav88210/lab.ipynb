```python
def get_rating_class(ratings):
    ratings_class = [] 
    for rating in ratings:
        rating_text = ''
        if rating <= 2:
            rating_text = 'Низкий рейтинг'
        elif rating > 2 and rating <= 4:
            rating_text = 'Средний рейтинг'
        elif rating > 4 and rating <= 5:
            rating_text = 'Высокий рейтинг'
        ratings_class.append(rating_text)
    return ratings_class

df = pd.read_csv('/Users/stanislav/Downloads/ratings.csv')
df['class'] = get_rating_class(list(df.rating))

print(df)
```

```python
from datetime import timedelta

def convert_timestamp_to_days(timestamps):
    list_days = []
    for timestamp in timestamps:
        days = timedelta(seconds=timestamp).days
        list_days.append(days)
    return list_days
        
def main():
    df = pd.read_csv('/Users/stanislav/Downloads/ratings.csv')


    # Выборка userId пользователей, которые поставили 100 и более рейтингов
    ratings_100_or_more = df.groupby(['userId']).size().reset_index(name='counts').query('counts >= 100')

    
    # выборка всех пользователей, где userId есть в списке переменной ratings_100_or_more
    records = df[ df['userId'].isin(list(ratings_100_or_more['userId'])) ]

    # Добавить доп. поле дни
    records['Lifetime'] = convert_timestamp_to_days(records['timestamp'])

    # подсчитать среднее значение 
    records = records.groupby(['userId']).agg({'Lifetime': ['mean']})

    print(records)
    
    

if __name__ == '__main__':
    main()
```

```python
from datetime import datetime

def production_year(list_timestamp, list_years):
    years = []
    for timestamp in list_timestamp:
        year = datetime.fromtimestamp(timestamp).year
        # 2 - для каждой строки пройдите по всем годам списка years
        if year in list_years:
            years.append(year) # 2 - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
        else:
            years.append(1900) # 2 -  если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
    return years


def main():
    films = pd.read_csv('/Users/stanislav/Downloads/films.csv')
    ratings = pd.read_csv('/Users/stanislav/Downloads/ratings.csv')

    #1. В переменную years запишите список из всех годов с 1950 по 2010.
    years = [year for year in range(1950, 2011)]

    # Объеделение таблиц
    films_and_ratings = films.merge(ratings, left_on=['movieId', 'movieId'], right_on=['userId', 'userId'], how='left')

    # 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'
    films_and_ratings['year'] = production_year(list(films_and_ratings['timestamp_x']), years)

    # 4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга
    films_and_ratings = films_and_ratings.groupby(['year']).agg({'rating': ['mean']}).sort_values([('rating', 'mean')],ascending=[False])
    print(films_and_ratings)

if __name__ == '__main__':
    main()

```
