```python
import pandas as pd

def get_rating_class(ratings):
    ratings_class = [] 
    for rating in ratings:
        rating_text = ''
        if rating <= 2:
            rating_text = 'Низкий рейтинг'
        elif rating > 2 and rating <= 4:
            rating_text = 'Средний рейтинг'
        elif rating > 4 and rating <= 5:
            rating_text = 'Высокий рейтинг'
        ratings_class.append(rating_text)
    return ratings_class

df = pd.read_csv('/Users/stanislav/Downloads/ratings.csv')
df['class'] = get_rating_class(list(df.rating))

print(df)
```
